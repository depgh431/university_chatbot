```python
import json
f=open('newdata.json')
data=json.load(f)
```

```python
l=[]
total=[]
for x in data.keys():
    for y in data[x]['Patterns']:
        l.append(y)
    total.append(l)
    l=[]
```

```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer
input_sentence="this is a test sentence, I wonder if it'll work"

model = BartForConditionalGeneration.from_pretrained('eugenesiow/bart-paraphrase')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to("cpu")
tokenizer = BartTokenizer.from_pretrained('eugenesiow/bart-paraphrase')
```

```python
total[0]
```

```python
new=[]
paraphrased=[]
for x in total:
    for y in x:
        input_sentence=y
        batch = tokenizer(input_sentence, return_tensors='pt')
        generated_ids = model.generate(batch['input_ids'],max_new_tokens=40)
        generated_sentence = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
        new.append(generated_sentence)
    paraphrased.append(new)
    new=[]
```

```python
j=0
l=[]
helo=[]
for x in data.keys():
    for y in data[x]['Patterns']:
        l.append(y)
    for k in paraphrased[j]:
        l.append(k[0])
    j+=1
    helo.append(l)
    l=[]
```

```python
pattern=helo
responses=[]
for x in data.keys():
    for y in data[x]['Responses']:
        l.append(y)
    responses.append(l)
    l=[]
```

```python
big_boi=dict()
for x in range(0,len(pattern)):
    big_boi[x]={"Patterns":pattern[x],"Responses":responses[x]}
```

```python
with open("paraphrased.json","w") as jsonfile:
    json.dump(big_boi,jsonfile)
```
