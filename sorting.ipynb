```python
import pandas as pd
data=pd.read_excel('final.xlsx')
df=pd.DataFrame(data)
```

```python
df
```

```python
l=[]
clusters=[]
for y in range(0,9):    
    clusters.append(0)
    for x in range(df.shape[0]):
        if df['cluster'][x]==y:
            l.append(df.iloc[x])
    clusters[y]=pd.DataFrame(l)
    l=[]
```

```python
clusters
```

```python
pattern=[]
responses=[]
for x in range(len(clusters)):
    pattern.append(clusters[x]['Queries'])
    responses.append(clusters[x]['Answers'])
```

```python
big_boi=dict()
for x in range(0,9):
    big_boi[x]={"Patterns":pattern[x].to_list(),"Responses":responses[x].to_list()}
```

```python
import json
with open('ourdata.json', 'w') as jsonfile:
    json.dump(big_boi, jsonfile)
```
